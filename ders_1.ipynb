# **Linux üzerinde Bash Komutları ile Veri Seti Keşfi ve Ön İşleme-1**
# **wget** komutuyla veri setimizi netten indirelim (örnek veri seti iris)

```python
!wget https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data
```

# **cat** komutuyla verisetimizi görelim

```python
cat iris.data
```

# **tac**  komutuyla veri setimizi ters sıralı (reverse) olarak görüntüleyelim

```python
!tac iris.data
```

### Ters (Reserve) olarak sıraladığımızda ilk satırın boş olduğunu tespit ettik. Bu satırı  **sed** komutu ile silelim.

```python
!sed -i '$d' iris.data
```

```python
!tac iris.data
```

# **head** komutuyla verisetimizin ilk 10 satırını görüntüleyelim

```python
!head iris.data # yada cat iris.data | head
```

# **head -n** ile istediğimiz sayıda veri satırı görüntüleyebiliriz.

```python
! head -n 15 iris.data # yada cat iris.data | head -n 15
```

# **tail (-n)** ile verisetimizin sondan istediğimiz sayıda satırını görüntüleyebiliriz.

```python
! tail iris.data # yada cat iris.data | tail
```

```python
! tail -n 4 iris.data # yada cat iris.data | tail -n 4
```

# **wc -l** komutuyla verisetimizin satır sayısını görüntüleyebiliriz.

```python
! wc -l iris.data # yada cat iris.data | wc -l
```

# **cut -d ',' -f** ile istediğimiz **sütunları** seçebiliriz, dilersek farklı bir isimle seçtiğimiz sütunları kaydedebiliriz. **-d** verilerin hangi ayraç ile birbirinden ayrıldığını belirtiyor.

```python
!cut -d ',' -f 1,2 iris.data  | head
```

```python
!cut -d ',' -f 1,2 iris.data  > iris2.csv
```

```python
! cat iris2.csv
```

# **tr** komutuyla verisetimizdeki ayracı 'delimiter veya seperator' değiştirebilir, dilersek farklı adla kaydedebiliriz.

```python
!tr ',' ';' < iris.data
```

```python
!tr ',' ';' < iris.data > iris-2.csv
```

# **sort -k** komutuyla istediğimiz sütunlara göre verisetimizi sıralayabiliriz. **-r** parametresiyle ters sıralama da yapabiliriz.

```python
!sort -k3 -k4 < iris.data
```

```python
!sort -r -k3 -k4 < iris.data
```

# **-n** numerik sort için

```python
!sort -n -k3 -k4 <iris.data | head
```

```python
!sort -r -n -k3 -k4 <iris.data | head
```

# **shuf -n** ile verisetimizden istediğimiz sayıda rastgele örnek alabiliriz.

```python
!shuf -n 22 iris.data
```

```python
!shuf -n 22 iris.data >iris-rastgele22.csv
```
